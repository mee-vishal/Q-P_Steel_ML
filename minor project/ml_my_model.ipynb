```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# for ploting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Splitting the dataset into the Training set and Test set

X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.3, random_state=0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score, make_scorer

classifier_knn = KNeighborsClassifier()
classifier_knn.fit(X_train,y_train)

# predicting the test set results
y_pred_knn=classifier_knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
y_pred_train_knn = classifier_knn.predict(X_train)
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_knn)))

# print the scores on training and test set
print('Training set score: {:.4f}'.format(classifier_knn.score(X_train, y_train)))
print(f"Training set f1-score; f1_score ={f1_score(y_train, y_pred_train_knn, average='macro')*100:0.1f}")
print('Test set score: {:.4f}'.format(classifier_knn.score(X_test, y_test)))
print(f"Test set f1-score; f1_score ={f1_score(y_test, y_pred_knn, average='macro')*100:0.1f}")
```

```python
from sklearn.model_selection import GridSearchCV
    #max_neighbors
param_dist = {
        "n_neighbors": [5,7,9,11,13,15],
        "weights": ['uniform','distance'],
        "algorithm": ['ball_tree', 'kd_tree', 'brute'],
        "leaf_size": [15,20,25,30],
        #"p":[1,2],
    }
rsh = GridSearchCV(
        estimator=classifier_knn, param_grid=param_dist)
rsh.fit(X_train, y_train)
rsh.best_params_
```

```python
classifier_knn = KNeighborsClassifier(n_neighbors = 5,
                                      weights ='distance',
                                      leaf_size = 15,
                                      p = 2,  # default value
                                      algorithm = 'ball_tree')
classifier_knn.fit(X_train,y_train)

# predicting the test set results
y_pred_knn=classifier_knn.predict(X_test)

y_pred_train_knn = classifier_knn.predict(X_train)
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_knn)))
# print the scores on training and test set

print('Training set score: {:.4f}'.format(classifier_knn.score(X_train, y_train)))
print(f"Training set f1-score; f1_score ={f1_score(y_train, y_pred_train_knn, average='macro')*100:0.1f}")
print('Test set score: {:.4f}'.format(classifier_knn.score(X_test, y_test)))
print(f"Test set f1-score; f1_score ={f1_score(y_test, y_pred_knn, average='macro')*100:0.1f}")
```

```python

def feature_scale(C, Si, Mn, TMAE, Ac1, Ac3, MS, Qtemp, Ptemp):

    # Mean values
    mean = {
        "C": 0.293711,
        "Si": 1.353189,
        "Mn": 2.130231,
        "TMAE": 0.017715,
        "Ac1": 742.445838,
        "Ac3": 829.178609,
        "MS": 357.581143,
        "Qtemp": 226.697143,
        "Ptemp": 367.342857
    }

    # Standard deviation values
    std = {
        "C": 0.175460,
        "Si": 0.538275,
        "Mn": 1.180975,
        "TMAE": 0.045991,
        "Ac1": 12.901875,
        "Ac3": 35.662434,
        "MS": 61.784294,
        "Qtemp": 77.922007,
        "Ptemp": 80.983794
    }

    # Feature scaling using (x-u)/sigma
    C_scaled = (C - mean["C"]) / std["C"]
    Si_scaled = (Si - mean["Si"]) / std["Si"]
    Mn_scaled = (Mn - mean["Mn"]) / std["Mn"]
    TMAE_scaled = (TMAE - mean["TMAE"]) / std["TMAE"]
    Ac1_scaled = (Ac1 - mean["Ac1"]) / std["Ac1"]
    Ac3_scaled = (Ac3 - mean["Ac3"]) / std["Ac3"]
    MS_scaled = (MS - mean["MS"]) / std["MS"]
    Qtemp_scaled = (Qtemp - mean["Qtemp"]) / std["Qtemp"]
    Ptemp_scaled = (Ptemp - mean["Ptemp"]) / std["Ptemp"]

    return [[
        C_scaled,
        Si_scaled,
        Mn_scaled,
        TMAE_scaled,
        Ac1_scaled,
        Ac3_scaled,
        MS_scaled,
        Qtemp_scaled,
        Ptemp_scaled
    ]]
```

```python
def decode_microstructure(predicted_value):

    mapping = {
        0: 'F,C',
        1: 'M,B,RA',
        2: 'M,B,RA,C',
        3: 'M,F,B,RA',
        4: 'M,F,RA',
        5: 'M,F,RA,C',
        6: 'M,RA',
        7: 'M,RA',
        8: 'M,RA,C'
    }

    return mapping[predicted_value]
```

```python
feature_cols = [ 'C', 'Si', 'Mn', 'TMAE', 'Ac1', 'Ac3', 'MS', 'Qtemp', 'Ptemp' ]
```

```python


def get_microstructure_ranges(microstructure):
    filtered = df[df["MicrostructureType"] == microstructure]

    if filtered.empty:
        print(f"No data found for {microstructure}")
        return

    print(f"\nRecommended Parameter Ranges for: {microstructure}\n")

    for col in feature_cols:

        min_val = round(filtered[col].min(), 2)
        max_val = round(filtered[col].max(), 2)
        mean_val = round(filtered[col].mean(), 2)

        print(f"{col}")
        print(f"Min  : {min_val}")
        print(f"Max  : {max_val}")
        print(f"Mean : {mean_val}")
```
